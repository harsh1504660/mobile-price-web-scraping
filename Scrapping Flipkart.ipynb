```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
import numpy as np
```

```python
name = []
rating = []
n_review = []
n_rating = []
ram=[]
rom=[]
expandable = []
price =[]
display=[]
battery = []
camera=[]

for j in range(1,50):
    
    webpage=requests.get('https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=offpage%3D5&page={}'.format(j)).text
    url = 'https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=offpage%3D5&page={}'.format(j)
    soup=BeautifulSoup(webpage,'lxml')
    product=soup.find_all(class_='_2kHMtA')
    
    for k in product:

        try:

            rating.append(float(k.find(class_='_3LWZlK').text))
        except:
            rating.append(np.nan)
        try:

            name.append(k.find(class_='_4rR01T').text)
        except:
            name.append(np.nan)

        try:

            n_rating.append(k.find(class_='_2_R_DZ').text.split()[0])
        except:
            n_rating.append(np.nan)

        try:

           
            n_review.append(k.find(class_='_2_R_DZ').text.split()[3])
        except:
              n_review.append(np.nan)
        try:

         
            ram.append(k.find(class_='rgWa7D').text.split()[0])
        except:
            ram.append(np.nan)

        try:


            rom.append(k.find(class_='rgWa7D').text.split()[4])
        except:
            rom.append(np.nan)
        try:

            expandable.append(k.find(class_='rgWa7D').text.split()[-2])
        except:
            expandable.append(np.nan)

        try:

            price.append(k.find(class_='_30jeq3 _1_WHN1').text.split()) 
        except:
            price.append(np.nan)

        try:
       
            display.append(k.find_all(class_='rgWa7D')[1].text.split()[2])
        except:
            display.append(np.nan)

        try:

            camera.append(k.find_all(class_='rgWa7D')[2].text.split()[0])
        except:
            camera.append(np.nan)

        try:

            battery.append(k.find_all(class_='rgWa7D')[3].text.split()[0])
        except:
            battery.append(np.nan)
        print(url)
```

```python
ncolumns = {'name':name,'ratings':rating,'number of rating':n_rating,'number of reviews':n_review,'ram':ram,'rom':rom,'expandable':expandable,'display':display,'camera':camera,'battery':battery,'price':price}
```

```python
new_data = pd.DataFrame(ncolumns)
```
